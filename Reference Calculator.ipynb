```python
from PIL import Image
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img_size = (800,600)
```

```python
img = Image.open("Images/car1.PNG", 'r')
img = img.resize(img_size)
img_gray = img.convert('L')
img_2 = Image.open("Images/test_segmented_car1.PNG", 'r')
img_2 = img_2.resize(img_size)
img_2 = img_2.convert('L')
```

```python
pix_val = np.array(img.getdata(),dtype=np.float64)
pix_val_gray = np.array(img_gray.getdata(),dtype=np.float64)
pix_2_val = np.array(img_2.getdata(),dtype=np.float64)
```

```python
def similarity_gen(A, B):
    """
    This method compares the average percent error between each pixel from the grayscale original image to the
    image to be compared against.
    Parameters:
    1. A: The original image (np.array, shape: n x 1).
    2. B: The image to be compared (np.array, shape: n x 1).
    Returns the average percent error of the difference in the test images' pixel value.
    """
    ols = 0
    for a,b in zip(A,B):
        ols += (a-b)**2
    return ols**0.5
```

```python
similarity_gen(pix_val_gray, pix_2_val)
```
